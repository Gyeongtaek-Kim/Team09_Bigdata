```python
#criticì˜ ì ìˆ˜ë¶„í¬ bar graphë¡œ í™•ì¸.
game_count = []
game_count2 = []
game_count3 = []
score = []
for i in range (10):
    a = len(game_meta_best.where(game_meta_best['Critic_Score']>= (9-i)*10).dropna())
    b = len(game_meta_best2.where(game_meta_best2['Critic_Score']>= (9-i)*10).dropna())
    c = len(game_meta_best3.where(game_meta_best3['Critic_Score']>= (9-i)*10).dropna())
    game_count.append(a)
    game_count2.append(b)
    game_count3.append(c)
    game_count[i] = game_count[i] - sum(game_count[0:i])
    game_count2[i] = game_count2[i] - sum(game_count2[0:i])
    game_count3[i] = game_count3[i] - sum(game_count3[0:i])
    score.append((9-i)*10)

print(game_count)
print(game_count2)
print(game_count3)
```

```python
game_count4 = []
game_count5 = []
game_count6 = []
score = []
for i in range (10):
    a = len(game_meta_worst.where(game_meta_worst['Critic_Score']>= (9-i)*10).dropna())
    b = len(game_meta_worst2.where(game_meta_worst2['Critic_Score']>= (9-i)*10).dropna())
    c = len(game_meta_worst3.where(game_meta_worst3['Critic_Score']>= (9-i)*10).dropna())
    game_count4.append(a)
    game_count5.append(b)
    game_count6.append(c)
    game_count4[i] = game_count4[i] - sum(game_count4[0:i])
    game_count5[i] = game_count5[i] - sum(game_count5[0:i])
    game_count6[i] = game_count6[i] - sum(game_count6[0:i])
    score.append((9-i)*10)

print(game_count4)
print(game_count5)
print(game_count6)
df = pd.DataFrame({"top 1000":game_count,"top 1000 to 2000":game_count2,"top 2000 to 3000":game_count3,
                  "worst 2000 to 3000":game_count4, "worst 1000 to 2000":game_count5, "worst 1000":game_count6})
ax = df.plot(kind = 'bar', figsize = (13,8), title = 'Score Distribution for each Rank Tier');
ax.set_xticklabels(["over 90","90~80","80~70","70~60","60~50","50~40","40~30","30~20","20~10","10~0"])
ax.set_xlabel('Critic Score')
ax.set_ylabel('Number of game')
```
