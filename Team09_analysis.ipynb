```python
#Question number 2. IGN과 메타크리틱, 두 게임 평론은 어떻게 다른가? 비슷하다면 얼마나 비슷한가?

#분석 방법 - 이는 모든 데이터가 기록된 merged에서 메타크리틱과 ign의 차이를 보면 될듯
#두 데이터를 하나의 테이블에 산포도로 나타내어 그 차이를 조사해 본다
#또한, user data와 크리틱 데이터 사이의 차이를 조사하여 유의미한 차이가 있는지 알아본다.
```

```python
#두 자료 merge
merged_game = pd.merge(game_ign, game_meta_valid, left_on='title', right_on='4')
merge_columns = merged_game.columns.tolist()
column_order = ['title', '4', 'genre', 'Genre', 'platform', 'Platform', 'score', 'score_phrase', 
                'editors_choice', 'Critic_Score', 'User_Score', 'Year_of_Release', 'release_year', 'release_month', 'release_day', 'Publisher',
                'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales'] #데이터프레임의 출력될 column의 순서

merge_result = merged_game[column_order]
merge_result = pd.DataFrame(merge_result.sort_values(by = 'Global_Sales', ascending=False))
```

```python
same_plat_compare = merge_result.where(merge_result['platform'] == merge_result['Platform']).dropna()
same_plat_compare = same_plat_compare.reset_index().reset_index()
same_plat_compare
```

```python
##metacritic 평론 결과와 ign 평론 결과를 비교 
## 양수 = ign이 더 높이 평가
## 음수 = metacritic critic들이 더 높이 평가
a = same_plat_compare['score']*10 - same_plat_compare['Critic_Score'] # 100 scale로 ign을 변환
a = pd.DataFrame(a, columns=['difference']).reset_index()
a.plot(x = 'index', y = 'difference', kind = 'scatter',figsize = (20,8))
```

```python
b = a['difference'].where(a['difference'] > 0).count()
c = a['difference'].where(a['difference'] < 0).count()
print(b,c)
##전체적으로 ign쪽 점수가 높다.
```

```python
##명제 오류 판단
##잘 팔린다의 기분 : 데이터셋의 첫 구간(1~1000등)에 해당하면 잘 팔린다고 판단하기로.
##메타크리틱 스코어의 평론가 점수를 기준으로 명제를 판별해보았다.

##1. 90점 이상의 게임이라면 그 게임은 잘 팔리는 게임이다.
##2. 게임이 잘 팔리는 게임이라면 해당 게임은 90점 이상이다.

game_meta_valid_best = game_meta_valid.head(1000)
#90점 이상이면서 1000등 이내.
ninety_true = game_meta_valid_best['Critic_Score'].where(game_meta_valid_best['Critic_Score'] >= 90).count()
#90점 이상 모든 갯수
ninety_all = game_meta_valid['Critic_Score'].where(game_meta_valid['Critic_Score'] >= 90).count()
#90점 이상이면서 1000등 밖
ninety_false = ninety_all - ninety_true
#90점 이하면서 1000등 이내
under_true = 1000 - ninety_true
#90점 이하고 1000등 밖
under_all = len(game_meta_valid) - under_true - ninety_false - ninety_true 
print(ninety_true, ninety_false, under_true, under_all)
```

```python
game_meta_valid_best = game_meta_valid.head(1000)
#80점 이상이면서 1000등 이내.
eighty_true = game_meta_valid_best['Critic_Score'].where(game_meta_valid_best['Critic_Score'] >= 80).count()
#80점 이상 모든 갯수
eighty_all = game_meta_valid['Critic_Score'].where(game_meta_valid['Critic_Score'] >= 80).count()
#80점 이상이면서 1000등 밖
eighty_false = eighty_all - eighty_true
#80점 이하면서 1000등 이내
under_true = 1000 - eighty_true
#80점 이하고 1000등 밖
under_all = len(game_meta_valid) - under_true - eighty_false - eighty_true 
print(eighty_true, eighty_false, under_true, under_all)
```
